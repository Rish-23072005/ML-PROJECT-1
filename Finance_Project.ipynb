#Importing Libraries

```python
from pandas_datareader import data as pdr
import pandas as pd
import numpy as np
import datetime as dt
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
%matplotlib inline
```

#Data Section

```python
start = dt.datetime(2006,1,1)
end = dt.datetime(2016,1,1)
```

#Let's Figure out the pickle

```python
# Use a different data source, e.g., 'stooq' or 'iex'
BAC = pdr.DataReader('BAC', data_source='stooq', start=start, end=end)
#citigroup
C = pdr.DataReader('C', data_source='stooq', start=start, end=end)
#JPMorgan chase
JPM = pdr.DataReader('JPM', data_source='stooq', start=start, end=end)
#morgan stanley
MS = pdr.DataReader('MS', data_source='stooq', start=start, end=end)
#well Fargo.
WFC = pdr.DataReader('WFC', data_source='stooq', start=start, end=end)
```

```python
WFC
```

```python
# Plot the data
C['Close'].plot()
plt.show()
```

#list of the ticketsymbols in alphabetical order

```python
tickers = ['BAC', 'C', 'JPM', 'MS', 'WFC']
```

```python
tickers
```

#Concatenate bank and bank_stokcs

```python
bank_stokcs = pd.concat([BAC, C, JPM, MS, WFC], axis=1, keys=tickers)
```

```python
print(bank_stokcs)
```

#Set column levels

```python
bank_stokcs.columns.names = ['Bank Ticker', 'Stock Info']
```

```python
bank_stokcs.head()
```

#EDA

```python
for tick in tickers:
  print(tick, bank_stokcs[tick]['Close'].max())
```

#Empty dataframe

```python
returns = pd.DataFrame()
```

#can use pct_change() method on the close column to creste column representing above return value

```python
for tick in tickers:
  returns[tick+' Return'] = bank_stokcs[tick]['Close'].pct_change()
```
